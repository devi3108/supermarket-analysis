### 1. Which city with the highest average total sales?

```python
ssal_city = ssal.groupby('City')['Total'].mean().sort_values(ascending=False)
```

```python
ssal_city
```

```python
_ = sns.countplot(data=ssal, x='City')
```

The three cities had nearly the same average total sales. The highest average total sales were Yangon City. Then followed by Mandalay City. The lowest average total sales were Naypitaw City.

### 2. What is customer favorite type of payment, based on their total shopping?

```python
ssal_pay = ssal.groupby('Payment')['Total'].sum()
```

```python
_ = ssal_pay.plot.pie(autopct = '%1.1f%%', labels=None, legend=True)
```

Payment by cash and Ewallet is the most preferred. Payment with cash is the most preferred, then followed by payment via Ewallet. The payment method by credit card gets the lowest percentage.

### 3. What is the highest selling time?

```python
ssal_time = ssal.groupby('Time')['Total'].mean()
```

```python
ssal_time.sort_values(ascending=False).head(10).plot(kind='bar')
_ = plt.xticks(rotation='horizontal')
```

Here are the 10 times with the highest average total sales. The time with the highest average total sales is 16.20. The times with the highest sales are around 16.00, 14.00, 12.00, 19.00, and 20.00.

### 4. What is the total expenditure for both women and men in each city?

```python
_ = sns.catplot(data=ssal, x='Total', y='Gender', col='City')
```

In the Yangon City, women spend an average of 200.000 to 400.000 bath on shopping. Meanwhile, men spent an average of 0 to 200 bath. In Naypitaw City, both women and men spend the most from 0 to 200.000 bath. Likewise in Mandalay City, however, many female customers in Mandalay spend around 400.000 on shopping.

### 6. What products do both women and men like the most?

```python
plt.figure(figsize=(12,9))
proline = sns.barplot(x="Product line", y="Total", hue="Gender", data=ssal)
proline.set_ylabel('Total', size=20)
proline.set_xlabel('Product line', size=20)
plt.show()
```

Female customers spend their money on products of Home and lifestyle, food and beverages, and fashion accessories. While male prefer the product of health and beauty, and also sports and travel. For Electronic accessories, both women and men do not have a significant difference when spending money to buy this product.

### 7. What product line contributes the most ti the gross income in each city?

```python
sns.set(rc={'figure.figsize':(12,10)})
_ = sns.violinplot("Product line", "gross income", hue='City', data=ssal)
```

Health and beauty has the largest gross income probability in Yangon, then Naypitaw, and followed by Mandalay. Although it has a fairly low probability, Mandalay has the highest median among those cities.
For electronic accessories product, Yangon and Mandalay have the same gross income probability. Meanwhile, Naypitaw is under the two city but has the highest median value of gross income.
Home and lifestyle has almost the same probability in the three cities. Sports and travel have the same gross income probability in Yangon and Mandalay. But, for the Naypitaw City the probability is less under the two cities and have the same median value. Food and beverages has the highest probability value in Yangon and Mandalay. This is followed by Naypitaw, but with the highest median gross income value among the three cities. Mandalay has a high probability value for product x, but has the lowest median value. In contrast to Yangon and Naypitaw which have a fairly low gross income probability of the product.
